In the code `data=dataset[:,0]` and `target=dataset[:,1]`, the colon `:` is used for slicing. It means "select all rows". So, `dataset[:,0]` selects all rows and the first column (index 0), and `dataset[:,1]` selects all rows and the second column (index 1).

Here are a few more examples of slicing with the colon `:` in NumPy arrays:

*   `my_array[start:end]` : Selects elements from `start` up to (but not including) `end`.
*   `my_array[start:]` : Selects elements from `start` to the end of the array.
*   `my_array[:end]` : Selects elements from the beginning up to (but not including) `end`.
*   `my_array[:]` : Selects all elements in the array.
*   `my_array[row_start:row_end, col_start:col_end]` : Selects a slice of both rows and columns.

## Solution - Adding another Feature (Higher Dimension)

```python
x1=dataset[:,0].reshape(-1,1)
print('x1:',x1)
x2=np.power(x1,2)
print('x2:',x2)
```

put x1 and x2 together

```python
data_new=np.append(x1,x2,axis=1) #if using axis =0 we can put x1 abd x2 one after another
print(data_new)
```

```python
train_data,test_data,train_target,test_target=train_test_split(data_new,target,test_size=0.1)
```

```python
model=LinearRegression()
model.fit(train_data,train_target)
```

```python
print('coefs:',model.coef_)
print('c:',model.intercept_)
```

look at above result there are two coefficients .

```python
a1=model.coef_[0][0]
a2=model.coef_[0][1]
a0=model.intercept_[0]
```

```python
xvals=np.arange(-10,10)
yvals=a1*xvals+a2*np.power(xvals,2)+a0
```

```python
plt.scatter(data,target)
plt.plot(xvals,yvals,'r',label='Best Fit Curve')
plt.xlabel('Feature -X)')
plt.ylabel('Label -Y)')
plt.legend()
plt.show()
```

```python
import pandas as pd
dataset = pd.read_csv('/content/drive/MyDrive/none-linear-dataset.csv').values
print(dataset,len(dataset))
```

```python
data=dataset[:,0].reshape(-1,1)
target=dataset[:,1]
print('data:',data)
print('target:',target)
```

use poly object to do transfromation

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2,include_bias=False)
data_new=poly.fit_transform(data)
print('data_new:',data_new)
```

```python
model=LinearRegression()
model.fit(data_new,target)
```

```python
print('m:',model.coef_)
print('c:',model.intercept_)
```

```python
test_data=[[7.5,7.5**2]]
result=model.predict(test_data)
print('result:',result)
```
